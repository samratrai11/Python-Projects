```python
# Libraries used 
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)
```

```python
# Reading the dataset
df=pd.read_csv('movies.csv')
# Loading the dataset
df
```

```python
# importing numpy for calucaltion
import numpy as np

# Checking missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Cheking the data types

df.dtypes
```

```python
# Chceking for Outliers

df.boxplot(column=['gross'])
```

```python
# Dropping duplicate values

df.drop_duplicates()
```

```python
# Ordering the Data for better presentation
# Sorting the dataset by the descending gross value

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Plotting a linear regression line with scatter plot
# Linear relationship between 'gross' and 'budget' 
# Results show a positive slope which mean postive correlation
# Scatter points around the line indicate the variability of the data

sns.regplot(x="gross", y="budget", data=df,scatter_kws={"color": "green"}, line_kws={"color": "blue"})
```

```python
# Linear relationship between 'gross' and 'budget' 

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
sns.regplot(x="score", y="gross", data=df, scatter_kws={"color": "goldenrod"}, line_kws={"color": "red"})

# Show the plot
plt.show()
```

```python
# Generate a correlation matrix for all numerical columns
# Computes the Pearson correlation coefficients between all pairs of numerical columns in the DataFrame


df.corr(method ='pearson')
```

```python
# calculates the Kendall rank correlation coefficients between all pairs of numerical columns in the df

df.corr(method ='kendall')
```

```python
# calculates the Spearman rank correlation coefficients between all pairs of numerical columns in the df

df.corr(method ='spearman')
```

```python
# The colors in the heatmap represent the strength and direction of correlations
# cooler colors indicating negative correlations
# warmer colors indicating positive correlations 
# darker shades indicating stronger correlations

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
# Using the factorize method assigns a unique integer label to each distinct categorical value in a column


df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm')

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
